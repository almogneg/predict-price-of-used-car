```python

import time      # for testing use only
import os         # for testing use only

import bs4
from bs4 import BeautifulSoup  
import pandas as pd
import scipy as sc
import numpy as np
import requests 
import matplotlib as mpl
from matplotlib import pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
mpl.rcParams['figure.dpi'] = 300
```

```python

##function load site
def load_soup_object(url):
    page = requests.get(url)
    page
    soup = BeautifulSoup(page.content, 'html.parser')
    return soup
```

```python
year=[]
model=[]
make=[]
km=[]
accident=[]
owner=[]
use=[]
price=[]
color=[]
###############

############
soup=load_soup_object("https://www.truecar.com/used-cars-for-sale/listings/")##load site
               

for i in range(220):
    relevant=soup.findAll("div",attrs={"class":"flex w-full flex-col"})## all the cars
    for par in relevant:

        #color
        h5s=par.find(attrs={"data-test":"vehicleCardColors"})
        string_h5s = [tag.get_text() for tag in h5s]
        color1=string_h5s[1]
        color.append(color1)


        #price
        h4s=par.find(attrs={"data-test":"vehicleCardPricingBlockPrice"})
        string_h4s = [tag.get_text() for tag in h4s]
        price1=string_h4s[0]
        price.append(price1)


        #year
        year.append(par.find("span",attrs={"class":"vehicle-card-year text-xs"}).string)


        #accident&&owner&&use
        h3s=par.find(attrs={"data-test":"vehicleCardCondition"})
        string_h3s = [tag.get_text() for tag in h3s]
        arr=string_h3s[1].split(",")
        accident1=arr[0]
        owner1=arr[1]
        use1=arr[2]
        accident.append(accident1)
        owner.append(owner1)
        use.append(use1)



        #km
        h2s=par.find(attrs={"data-test":"vehicleMileage"})
        string_h2s = [tag.get_text() for tag in h2s]
        km.append(string_h2s[1])

        #make&&model
        h1s=par.find("span",attrs={"class":"truncate"})
        string_h1s = [tag.get_text() for tag in h1s]
        make.append(string_h1s[0])
        model.append(string_h1s[4])
        
    #next_page
    nextUrl=soup.find("a", attrs={"aria-label":"Go to next page"})['href']##url to the next page
    next_url="https://www.truecar.com"+nextUrl
    soup= load_soup_object(next_url)

df=pd.DataFrame({"year":year,"make":make,"model":model,"km":km,"accident":accident,"owner":owner,"use":use,"color":color,"price":price})
```
